```python
# Importer les données
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Fasonet_Commentaires_Article.csv')
df.drop(df.columns[0], axis=1,inplace=True)
df.head()
```

```python
# Fonction pour tronquer la colonne date_heure
def TroncateDate(data,colonne):
  #data['Date']=colonne
  for i in range(len(colonne)):
    data.iloc[i,3]=data.iloc[i,3][:10]
  return data

df=TroncateDate(df,df.Date_Heure)

df.head()
```

```python
#Fonction pour convertir en type date
def convertDate(colonne):
  colonne=pd.to_datetime(colonne)
  return df

# Fonction pour renommer la colonne date heure en date
def renameColonne(data):
  data.rename(columns = {'Date_Heure':'Date'}, inplace = True)
  return data

df=convertDate(df.Date_Heure)
df=renameColonne(df)

df.head()
```

```python
# Fonction pour supprimer les doublons dans les commentaires
def supprimerDoublon(data):
  #Avant suppression des doublons
  data.drop_duplicates(subset='Commentaire',inplace=True)
  return data

# Appel de la fonction
df=supprimerDoublon(df)
```

```python
# Fonction pour vérifier les valeurs manquantes
def verifierValeurManquante(data):
  NAN = [(c, data[c].isna().mean()*100) for c in data]
  NAN = pd.DataFrame(NAN, columns=["column_name", "percentage"])
  return(NAN.sort_values("percentage", ascending=False))

#Appel de la fonction
verifierValeurManquante(df)
```

**Nettoyage du jeux de données**

```python
# Fonction pour nettoyer le corpus
def nlp_pipeline(text):

    text = text.lower()
    text = text.replace('\n', ' ').replace('\r', '')
    text = ' '.join(text.split())
    text = re.sub(r"[A-Za-z\.]*[0-9]+[A-Za-z%°\.]*", "", text)
    text = re.sub(r"(\s\-\s|-$)", "", text)
    text = re.sub(r"[,\!\?\%\(\)\/\"]", "", text)
    text = re.sub(r"\&\S*\s", "", text)
    text = re.sub(r"\&", "", text)
    text = re.sub(r"\+", "", text)
    text = re.sub(r"\#", "", text)
    text = re.sub(r"\$", "", text)
    text = re.sub(r"\£", "", text)
    text = re.sub(r"\%", "", text)
    text = re.sub(r"\:", "", text)
    text = re.sub(r"\@", "", text)
    text = re.sub(r"\-", "", text)
    return text

# Appel de la fonction
df['Commentaire_clean'] = df['Commentaire'].apply(nlp_pipeline)
```

```python
#Fonction pour la tokénisation
def tokenisationCorpus(data,colonne):
  AComment=[]
  for comment in colonne.apply(str):
      Word_Tok = []
      for word in  re.sub("\W"," ",comment ).split():
          Word_Tok.append(word)
      AComment.append(Word_Tok)
  data["Word_Tok"]= AComment
  return data

# Appel de la fonction
df=tokenisationCorpus(df,df.Commentaire_clean)
```

```python
#Fonction pour la suppréssion des stopwords
def supprimerStopWords(data,colonne):
  stop_words=set(STOP_WORDS)
  deselect_stop_words = ['n\'', 'ne','pas','plus','personne','aucun','ni','aucune','rien']
  for w in deselect_stop_words:
      if w in stop_words:
          stop_words.remove(w)
      else:
          continue
  AllfilteredComment=[]
  for comment in colonne:
      filteredComment = [w for w in comment if not ((w in stop_words) or (len(w) == 1))]
      AllfilteredComment.append(' '.join(filteredComment))
  data["CommentaireSansMotVide"]=AllfilteredComment
  return(data)

# Appel de la fonction supprimerStopWords()
df=supprimerStopWords(df,df.Word_Tok)
```

**Analyse des sentiments avec Textblob**

```python
# Fonction pour la polarisation
def sentimentAnalysis(data,colonne):
  senti_list = []
  for i in colonne:
      vs = tb(i).sentiment[0]
      if (vs > 0):
          senti_list.append('Positive')
      elif (vs < 0):
          senti_list.append('Negative')
      else:
          senti_list.append('Neutral')
  data["Sentiment"]=senti_list
  return data

df=sentimentAnalysis(df,df.CommentaireSansMotVide)
```

```python
# Fonction pour ajouter les stopwords à supprimer
def motsVides():
  stop_words=set(STOP_WORDS)
  for i in ['hum','hey','walai','qu']:
    stop_words.add(i)
  return stop_words
```

```python
#Fonction pour la suppréssion des stopwords
def supprimerStopWords2(data,colonne):
  stop_words=motsVides()
  AllfilteredComment=[]
  for comment in colonne:
      filteredComment = [w for w in comment if not ((w in stop_words) or (len(w) == 1))]
      AllfilteredComment.append(' '.join(filteredComment))
  data["CommentaireSansMotVide"]=AllfilteredComment
  return(data)

# Appel de la fonction supprimerStopWords()
df_Thematique=supprimerStopWords2(df_Thematique,df_Thematique.Word_Tok)
```

```python
# Lemmatization
nlp = spacy.load("fr_core_news_md")
df["Commentaire_lem"]=df["CommentaireSansMotVide"].apply(lambda row: " ".join([w.lemma_ for w in nlp(row)]))
```

```python
def lemmatization(texts,allowed_postags=["Nouns","ADJ","VERB","ADV"]):
    nlp=spacy.load("fr_core_news_md",disable=["parser","ner"])
    texts_out=[]
    for text in texts:
        doc=nlp(text)
        new_text=[]
        for token in doc:
            if token.pos_ in allowed_postags:
              new_text.append(token.lemma_)
        final=" ".join(new_text)
        texts_out.append(final)
    return(texts_out)

df_Thematique["Commentaire_lem"]=lemmatization(df_Thematique["CommentaireSansMotVide"])
```

```python
# Fonction permettant de compter le nombre de mot dans le corpus
def nombreMots():
  all_text = ' '.join([text for text in df_Thematique['Commentaire_lem']])
  print('Number of words in all_text:', len(all_text))

nombreMots()
```

```python
# Fonction permettant d'afficher l'histogramme des termes dans le corpus
def histoFreqCorpus():
  # un Barplot avec le compte des nombreTerme premiers mots les plus fréquents dans le corpus
  freq = ' '.join(text for text in df_Thematique['Commentaire_lem'])
  list = freq.split()
  counts = Counter(list)
  common_words = [word[0] for word in counts.most_common(20)]
  common_counts = [word[1] for word in counts.most_common(20)]
  #plt.style.use('dark_background')
  plt.figure(figsize=(15, 12))
  colors_list =['Red','Orange', 'Blue', 'Purple']
  graph=plt.bar(common_words,common_counts,color=colors_list)
  plt.title('Les {} premiers termes les plus utilisés dans le corpus'.format(20));
  i = 0
  for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
        y+height*1.01,
        str(common_counts[i])+'',
        ha='center',
        weight='bold')
    i+=1
  plt.show()

histoFreqCorpus()
```

```python
import numpy as np
import json
import glob

#NLTK
import nltk

#Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel

#Spacy
import spacy
from nltk.corpus import stopwords

#vis
import pyLDAvis
import pyLDAvis.gensim_models #as gensimvis
pyLDAvis.enable_notebook()

import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)

# Changer de répertoire par defaut
#import os
#os.chdir("le chemin")
```

```python
# Récuperer sous forme de liste
corpus=df['Commentaire_lem'].tolist()
print(corpus[0])
```

```python
# Transformer le corpus en liste de listes (les documents) par tokénisation
#nltk.download('punkt')
#from nltk.tokenize import word_tokenize
tokenizer = nltk.RegexpTokenizer(r'\w+') # tokenizer spécifique au francais
corpus_tk=[tokenizer.tokenize(doc) for doc in corpus]

#Avant tokenisation
print(corpus[0])

# Après tokenisation
print('\n')
print(corpus_tk[0])
```

```python
#Charger les stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
mots_vides=stopwords.words('french')

# Compléter la liste des stopwords avec des termes propres à notre corpus
mots_vides.extend(['hum','hey','walai','être','avoir','tout','faire','comme','falloir'])
```

```python
#Suppression des stops words ou mots vides
corpus_sw=[[mot for mot in doc if not mot in mots_vides] for doc in corpus_tk]

# Verification origine
print(corpus_tk[0])

# Verification sans les stopswords
print(corpus_sw[0])
```

```python
# Rétirer les tokens de moins de 3 lettres
corpus_sw=[[mot for mot in doc if len(mot)>3] for doc in corpus_sw]
print(corpus_sw[0])
```

```python
corpus_sentence=[]
for liste_token in corpus_sw:
  corpus_sentence.append(' '.join(liste_token))
```

```python
#La lemmatization
nlp = spacy.load("fr_core_news_md")
#df["Commentaire"]=df["Commentaire"].apply(lambda row: " ".join([w.lemma_ for w in nlp(row)]))
corpus_lm=[[mot.lemma_ for mot in nlp(doc)] for doc in corpus_sentence]
```

```python
# Construction d'un dictionnaire de tokens
from gensim.corpora import Dictionary
dico=Dictionary(corpus_lm)
print(dico)
```

```python
#Liste des tokens et leur numéro interne
print(dico.token2id)
```

```python
#Representation bag of words des documents
corpus_bow=[dico.doc2bow(doc) for doc in corpus_lm]

#Verification
print(sorted(corpus_lm[0]))

#et en bow
print('\n')
print(corpus_bow[0])
```

```python
# Topic Modeling avec l'algorithme LDA
from gensim.models import LdaModel
lda=gensim.models.ldamodel.LdaModel(corpus=corpus_bow,
             id2word=dico,
             num_topics=5,
             random_state=100,
             passes=15,
             update_every=1,chunksize=100,
             alpha='auto',
             iterations=0,
             per_word_topics=True)
```

```python
# Detail sur le premier topic
print(lda.show_topic(0))
```

```python
pyLDAvis.enable_notebook()
vis=pyLDAvis.gensim_models.prepare(lda,corpus_bow,dico,mds="mds",R=30)
vis
```
